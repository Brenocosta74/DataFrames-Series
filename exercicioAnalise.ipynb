```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("dados_matriculas.xlsx")

```

```python
# Grafico de matriculas mensais

# Tiutlo, leganda, pontinho , eixo x, eixo y

plt.figure(figsize=(10,5))
df['Matrículas'].plot(marker='D', linestyle='-', color='blue', mfc='purple', ms='15')
plt.title('Mátriculas Mensais')
plt.xlabel('Mês')
plt.ylabel('Números de Matriculas')

#Mostrar apenas o gráfico
plt.show()

#Mostrar a tabela com o gráfico
#display(df)
```

```python
# Identificar o mês com o maior e menor número de matriculas

maior_matricula = df['Matrículas'].max()
indice_maior_matricula = df['Matrículas'].idxmax()

display(maior_matricula)
display(indice_maior_matricula)
```

```python
menor_matricula = df['Matrículas'].min()
indice_menor_matricula = df['Matrículas'].idxmin()

display(menor_matricula)
print(indice_menor_matricula)
```

```python
plt.figure(figsize=(10,5))
df['Matrículas'].plot(marker='D', linestyle='-', color='blue', mfc='white', ms='10', label='Matriculas')
plt.axhline(y=maior_matricula,
            color = "Black",
            linestyle = ":",
            linewidth=2,
            label=f"Máximo: {maior_matricula}")

plt.title('Matrículas Mensais')
plt.xlabel('Mês')
plt.ylabel('Número Matriculas')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import pandas as pd

df = pd.read_excel('dados_matriculas.xlsx')

img = plt.imread('richarlison.jpg')

fig, ax = plt.subplots()

ax.plot(df['Matrículas'], linestyle='-')  

for i, matricula in enumerate(df['Matrículas']):
    imagebox = OffsetImage(img, zoom=0.18)
    ab = AnnotationBbox(imagebox, (i, matricula), frameon=False)
    ax.add_artist(ab)

plt.show()
```
